In this program we write a simple gui using pythons tkinter package for a potential learning management system. This program is written for the user. When the user inputs the information for each student and submitted, it is then stored in a txt/csv file.  

```python
import tkinter as tk
from tkinter.filedialog import askopenfilename
import pandas as pd
from tkinter import filedialog
from tkinter import *
import csv
import tkinter.simpledialog
```

```python
#Storing the variables
def store():
    
    aname = name.get()
    astdID = stdID.get()
    acourse1 = course1.get()
    acourse2 = course2.get()
    acourse3 = course3.get()
    acourse4 = course4.get()
    aexam1 = exam1.get()
    aexam2 = exam2.get()
    aexam3 = exam3.get()
    aexam4 = exam4.get()
    agpa = gpa.get()
    
    #Checking for errors
    
    if(aname == "" or astdID == "" or agpa == ""):
        print("Error")
        tk.messagebox.showerror("error", "there was an issue with some information")
        name.set("")
        astdID.set()
        agpa.set()
    #Making sure that exam scores are not more than 100
    elif(aexam1 > 100):
        print("Error 2")
        tk.messagebox.showerror("error", "please enter a valid grade")
        name.set("")
        astdID.set()
        agpa.set()
    elif(aexam2 > 100):
        print("Error 2")
        tk.messagebox.showerror("error", "please enter a valid grade")
        name.set("")
        astdID.set()
        agpa.set()
    elif(aexam3 > 100):
        print("Error 2")
        tk.messagebox.showerror("error", "please enter a valid grade")
        name.set("")
        astdID.set()
        agpa.set()
    elif(aexam4 > 100):
        print("Error 2")
        tk.messagebox.showerror("error", "please enter a valid grade")
        name.set("")
        astdID.set()
        agpa.set()
    #Making sure that the gpa is valid
    elif(agpa > 4.0):
        print("Error 3")
        tk.messagebox.showerror("error", "please enter a valid GPA")
        name.set("")
        astdID.set()
        agpa.set()
    
    #If no errors program will ask user to submit
    else:
        result = tk.messagebox.askquestion("Submit", "You are about to enter the following data\n" + aname)
        
        #stores the values into a csv file
        if(result == 'yes'):
            print("here")
            #opens the students files
            with open('stdInfo.txt', 'a') as csvfile:
                writer = csv.writer(csvfile)
                writer.writerow([aname,astdID, acourse1, acourse2, acourse3, acourse4, aexam1, aexam2, aexam3, aexam4, agpa])
            csvfile.close()
        else:
            name.set("")
            astdID.set("")
            agpa.set("")
def back():
    print("here")

#clears all values 
def clear():
    name.set("")
    stdID.set(0)
    course1.set(0)
    course2.set(0)
    course3.set(0)
    course4.set(0)
    exam1.set(0)
    exam2.set(0)
    exam3.set(0)
    exam4.set(0)
    gpa.set(0)
    
#helps user asking for help    
def helpEntry():
    tk.messagebox.showinfo("Information", "Enter data for each student")
    
```

```python
root = tk.Tk()
#defining all variables
name = StringVar()
stdID = IntVar()
course1 = IntVar()
course2 = IntVar()
course3 = IntVar()
course4 = IntVar()
exam1 = IntVar()
exam2 = IntVar()
exam3 = IntVar()
exam4 = IntVar()
gpa = IntVar()

canvas = tk.Canvas(root,background = "light grey", height = 700, width = 800)
canvas.pack()

frame = tk.Frame(root, bg = '#80c1ff', bd = 5)
frame.place(relx = 0.5,rely = 0.1, relwidth =0.75,relheight = 0.1, anchor = 'n')

#label for student info
lblHeading = tk.Label(frame, text = "Student Info - User", font = ("Arial", 24, "bold"))
lblHeading.place(relwidth = 1, relheight = 1)

#lower frame
Lframe = tk.Frame(root, bg = "#80c1ff",bd = 15)
Lframe.place(relx = 0.5, rely = 0.25, relwidth = 0.75, relheight = 0.6, anchor = 'n')

#Creating labels and entry columns for the user to input
lblName = tk.Label(Lframe, text = "Name", font = ("Arial", 12, "bold"))
lblName.place(relx = 0.01, rely = 0.1)

entryName = tk.Entry(Lframe, textvariable = name, font = 30)
entryName.place(relx = 0.15, rely = 0.1)

lblstdID = tk.Label(Lframe, text = "Student ID", font = ("Arial", 12, "bold"))
lblstdID.place(relx = 0.5, rely = 0.1)

entrylblstdID = tk.Entry(Lframe,textvariable = stdID, font = 40)
entrylblstdID.place(relx = 0.7, rely = 0.1, relwidth = 0.2)

lblCourse1 = tk.Label(Lframe, text = "Course 1 ID", font = ("Arial", 12, "bold"))
lblCourse1.place(relx = 0.01, rely = 0.2)

entrylblCourse1 = tk.Entry(Lframe,textvariable = course1, font = 40)
entrylblCourse1.place(relx = 0.15, rely = 0.2)

lblCourse2 = tk.Label(Lframe, text = "Course 2 ID", font = ("Arial", 12, "bold"))
lblCourse2.place(relx = 0.01, rely = 0.3)

entrylblCourse2 = tk.Entry(Lframe,textvariable = course2, font = 40)
entrylblCourse2.place(relx = 0.15, rely = 0.3)


lblCourse3 = tk.Label(Lframe, text = "Course 3 ID", font = ("Arial", 12, "bold"))
lblCourse3.place(relx = 0.01, rely = 0.4)

entrylblCourse3 = tk.Entry(Lframe,textvariable = course3, font = 40)
entrylblCourse3.place(relx = 0.15, rely = 0.4)

lblCourse4 = tk.Label(Lframe, text = "Course 4 ID", font = ("Arial", 12, "bold"))
lblCourse4.place(relx = 0.01, rely = 0.5)

entrylblCourse4 = tk.Entry(Lframe,textvariable = course4, font = 40)
entrylblCourse4.place(relx = 0.15, rely = 0.5)


lblExam1 = tk.Label(Lframe, text = "Exam Score", font = ("Arial", 12, "bold"))
lblExam1.place(relx = 0.5, rely = 0.2)

entrylblExam1 = tk.Entry(Lframe,textvariable = exam1, font = 40)
entrylblExam1.place(relx = 0.7, rely = 0.2, relwidth = 0.2)

lblExam2 = tk.Label(Lframe, text = "Exam Score", font = ("Arial", 12, "bold"))
lblExam2.place(relx = 0.5, rely = 0.3)

entrylblExam2 = tk.Entry(Lframe, textvariable = exam2,font = 40)
entrylblExam2.place(relx = 0.7, rely = 0.3, relwidth = 0.2)

lblExam3 = tk.Label(Lframe, text = "Exam Score", font = ("Arial", 12, "bold"))
lblExam3.place(relx = 0.5, rely = 0.4)

entrylblExam3 = tk.Entry(Lframe, textvariable = exam3,font = 40)
entrylblExam3.place(relx = 0.7, rely = 0.4, relwidth = 0.2)

lblExam4 = tk.Label(Lframe, text = "Exam Score", font = ("Arial", 12, "bold"))
lblExam4.place(relx = 0.5, rely = 0.5)

entrylblExam4 = tk.Entry(Lframe,textvariable = exam4, font = 40)
entrylblExam4.place(relx = 0.7, rely = 0.5, relwidth = 0.2)

lblgpa = tk.Label(Lframe, text = "Current GPA", font = ("Arial", 12, "bold"))
lblgpa.place(relx = 0.25, rely = 0.7)

entrygpa = tk.Entry(Lframe,textvariable = gpa, font = 40)
entrygpa.place(relx = 0.45, rely = 0.7)


#Creating buttons
submitButton = tk.Button(Lframe, text = "Store", command = store)
submitButton.place(relx = 0.01, rely = 0.9)

backButton = tk.Button(Lframe, text = "Back", command = back)
backButton.place(relx = 0.30, rely = 0.9)

clearButton = tk.Button(Lframe, text = "Clear", command = clear)
clearButton.place(relx = 0.60, rely = 0.9)

helpButton = tk.Button(Lframe, text = "Help", command = helpEntry)
helpButton.place(relx = 0.90, rely = 0.9)



root.mainloop()
```
